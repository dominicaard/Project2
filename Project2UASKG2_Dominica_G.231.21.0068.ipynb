```python
import matplotlib.pyplot as plt

# Data
asimilasi = [23.3, 25.5, 25.6, 24.8, 22.5, 22.4, 22.95, 22.0, 25.8]
observasi = [24.2, 25.2, 23.9, 23.9, 24.0, 24.2, 24.0, 22.8, 23.0]
non_asimilasi = [23.6, 25.7, 25.3, 25.5, 24.9, 25.1, 24.7, 24.3, 25.9]

# Plot garis dengan titik
plt.plot(asimilasi, marker='o', label='Asimilasi', color='blue')
plt.plot(observasi, marker='o', label='Observasi', color='yellow')
plt.plot(non_asimilasi, marker='o', label='Non Asimilasi', color='red')

# Penyetingan plot
plt.xlabel('Data Ke-', weight='bold')
plt.ylabel('Suhu Permukaan', weight='bold')
plt.title('Visualisasi Data Suhu Permukaan', weight='bold')
plt.legend(loc='lower right', bbox_to_anchor=(1.35, 0))

# Tampilkan plot
plt.show()
```
